```python
import json
import graphviz

# Read the room definitions
with open('../src/Game/data/room-defs.json', 'r') as f:
    rooms = json.load(f)

# Create a new directed graph
dot = graphviz.Digraph(comment='Room Map')
dot.attr(rankdir='TB')

# Add all rooms as nodes
for room in rooms:
    dot.node(room['id'], room['name'])

# Add all exits as edges
for room in rooms:
    for exit in room['exits']:
        dot.edge(room['id'], exit['destinationId'], exit['direction'])

# Display the graph
dot
```
